```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("../data/weight-height.csv")
```

```python
dataframe.head()
```

```python
dataframe.describe()
```

```python
dataframe.plot(x = 'Height',y = 'Weight',kind = 'scatter')
```

```python
fig,ax = plt.subplots(1,2)
males = dataframe[dataframe['Gender'] == 'Male']
females = dataframe[dataframe['Gender'] == 'Female']
males.plot(kind = 'scatter',x = 'Height',y = 'Weight',ax = ax[0],color = 'blue',alpha = 0.3)
females.plot(kind = 'scatter',x = 'Height',y = 'Weight',ax = ax[1],color = 'red',alpha = 0.3)
```

```python
dataframe['Gendercolor'] = dataframe['Gender'].map({'Male':'blue','Female':'red'})
dataframe.plot(kind = 'scatter',x = 'Height', y = 'Weight',c = dataframe['Gendercolor'])
```

```python
males["Height"].plot(kind = 'hist',range = (50,80),bins = 50,title = 'Histogram',alpha =0.3,color = 'blue')
females["Height"].plot(kind = 'hist',range = (50,80),bins = 50,title = 'Histogram',alpha =0.3,color = 'red')
plt.axhline(y = males.Height.mean(),color = 'yellow')
plt.axhline(y = females.Height.mean(),color = 'black')
```
