# Visual Data Exporation Using MatplotLib

```python
"""Generating some random data using python"""
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data1 = np.random.normal(0,0.1,1000)
data2 = np.random.normal(1,0.4,1000) + np.linspace(0,1,1000)
data3 = 2 + np.random.random(1000) * np.linspace(0,5,1000)
data4 = np.random.normal(3,0.2,1000) + 0.3 * np.sin(np.linspace(0,20,1000))
```

```python
data = np.vstack([data1,data2,data3,data4]).transpose()
```

```python
data
```

```python
df = pd.DataFrame(data, columns =['data1','data2','data3','data4'] )
df.head()
```

# Line Plotting using Pandas and Matplotlib

```python
"""For Pandas,Line Plot is default."""
```

```python
df.plot(title = 'Line plot')
```

```python
plt.plot(df)
plt.title('Line Plot')
plt.legend(['data1','data2','data3','data4'])
```
