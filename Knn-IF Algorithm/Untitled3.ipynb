```python
distances = []
import operator
No_of_test_cases = X_test.shape[0]
No_of_test_cases
```

```python
def cluster_total(ClusterNeighboursList):
    ClusterTotal = 0
    for each_row in ClusterNeighboursList:
        x,y,z = each_row
        ClusterTotal += x+y+z
#     print(ClusterNeighbours1Total)
    return ClusterTotal
```

```python
def featuretotal(ClusterNeighbours,IndicesToFeature):
    FeatureTotalSum = 0
#     print(ClusterNeighbours)
    x,y,z = ClusterNeighbours[IndicesToFeature]
    FeatureTotalSum += x+y+z
#     print(ClusterNeighbours1Total)
    return FeatureTotalSum
```
